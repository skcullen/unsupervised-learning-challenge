```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Loading data
file_path = Path("crypto_data.csv")
df = pd.read_csv(file_path)
df
```

```python
#Get rid of all currencies that are currently trading
#drop unneeded column
df2 = df[df['IsTrading'] == True]
df3 = df2.drop(['IsTrading'], axis='columns')
df3
```

```python
#drop all rows that contain null values
df4 = df3.dropna()
df4
```

```python
#Get all values where data mined > 0
df5 = df4[df4['TotalCoinsMined'] > 0]
df5
```

```python
#Get rid of naming columns (non-numerical data that aren't catagorical)
df6 = df5.drop(['Unnamed: 0','CoinName'], axis='columns')
df6
```

```python
#Convert remaining non-numerical data into numerical using dummy variables in pandas
df7 = pd.get_dummies(df6)
df7
```

I have shown the dataframe at each step so that I could keep track of the rows and columns as the data was prepared.
We started with 1252 rows and 7 columns. 
All the preparation steps before we get the dummy variables have slowly whittled the set down to 532 rows and 4 columns.
Getting the dummy variables causes the columns to jump up to 377.

```python
#Scale the data using Standard Scaler
scaler = StandardScaler()
scaled_crypto = scaler.fit_transform(df7)
scaled_crypto
```

```python
#Perform dimensionality reduction with PCA (preserve 90% of the explained variance)
pca = PCA(n_components=0.90)
pca_transformed_crypto = pca.fit_transform(scaled_crypto)
pca_transformed_crypto
```

```python
#check difference in number of features before and after the pca
len(scaled_crypto[0])
```

```python
len(pca_transformed_crypto[0])
```

The number of feature drops (predictably), but not very drastically.

```python
#further reduce the dataset dimensions with t-SNE
tsne = TSNE(learning_rate=350)
tsne_features = tsne.fit_transform(pca_transformed_crypto)
tsne_features
```

```python
# check shape
tsne_features.shape
```

```python
#Visualize reduced data using scatterplot
# Get transformed features
df7['x'] = tsne_features[:,0]
df7['y'] = tsne_features[:,1]
# Bulid plot
plt.scatter(df7['x'], df7['y'])
plt.show()
```

There seems to be a huge cluster centering around (0,0), but there doesnt seem to be another defined cluster visable.
The outliers seem to form a ring around the big cluster. I have run it several times and it seems to go all the way around. That is definately a pattern, but not really a cluster.

```python
#Use a for-loop to determine the inertia for each k between 1 through 10.
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=11)
    km.fit(pca_transformed_crypto)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve for Cryptocurrency Data')
plt.show()
```

I ran this elbow curve for a couple of different random states(42 is particularly strange). There does not seem to be a good elbow for cryptocurrency.
I would advise a client that there is not a particularly good way to cluster cryptocurrency data. Most of it exists in a single cluster, which doesn't help us beyond determining outliers.
